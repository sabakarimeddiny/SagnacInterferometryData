```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
from analysis import *
import glob
import re
import hvplot.xarray
from scipy.optimize import curve_fit
font = {'family':'serif',
        'weight': 'normal',
        'size': 32}
plt.rc('font',**font)
import matplotlib as mpl
mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=['#000000','#e3130f','#0962f6',
                                                    '#20c168','#fc76f3','#2efef4'])
plt.rc('text',usetex=True)
```

```python
mp = pd.read_csv('test_SagnacPHE_F0.0900T_V5.0V_P0.0deg_2021-07-07_1.csv',comment='#')
mm = pd.read_csv('test_SagnacPHE_F0.0900T_V5.0V_P0.0deg_2021-07-07_2.csv',comment='#')
```

```python
plt.plot(mp.field_azimuth,mp.TY2)
plt.plot(mm.field_azimuth,mm.TY2)
```

```python
Rahe = 0.3548465783757197
Rphe = 0.2006307288681012
Meff = 0.44
R = 336.123 #ohms
```

```python
popt1
```

```python
popt2
```

```python
def mp_func(x,Hx,Hy,c):
    return ((Hx*Rahe - Hy*Rphe)*np.cos(np.pi/180*(x)) +\
            (Hy*Rahe - Hx*Rphe)*np.sin(np.pi/180*(x)))/2/Meff**2*(5/R)*0.09 + c
def mm_func(x,Hx,Hy,c):
    return -((-Hx*Rahe + Hy*Rphe)*np.cos(np.pi/180*(x)) +\
            (Hy*Rahe - Hx*Rphe)*np.sin(np.pi/180*(x)))/2/Meff**2*(5/R)*0.09 - c

popt1,pcov = curve_fit(mp_func,mp.field_azimuth.values,mp.TY2.values)
plt.plot(mp.field_azimuth,mp.TY2)
plt.plot(mp.field_azimuth,mp_func(mp.field_azimuth,*popt1))
popt2,pcov = curve_fit(mm_func,mm.field_azimuth.values,mm.TY2.values)
plt.plot(mm.field_azimuth,mm.TY2)
plt.plot(mm.field_azimuth,mm_func(mm.field_azimuth,*popt2))
```
